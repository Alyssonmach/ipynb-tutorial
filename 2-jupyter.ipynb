# O que é o Software Jupyter?
***
O Jupyter Notebook é um interfarce gráfica que permite a edição de notebooks em um navegador web, tais como Google Chrome ou Firefox. O nome Jupyter é um acrônimo criado a partir das linguagens de programação que inicialmente foram aceitas pelo Projeto Jupyter: Julia, Python e R.

## Interface

Quando você abre o `Jupyter Notebook` ou o `JupyterLab`, é possível escolher em qual diretório será criado um novo arquivo `.ipynb`. O processo é extremamente intuitivo, embora haja pequenas diferenças entre as duas versões do Jupyter.

![jupyter-notebook](assets/interface1.png)
> *O Jupyter Notebook permite navegação entre os diretórios, de modo que é possível especificar os caminhos em que os notebooks serão criados ou importados.*

![jupyter-lab](assets/interface2.png)
> *O JupyterLab também permite navegação entre os diretórios, de modo que é possível especificar os caminhos em que os notebooks serão criados ou importados*

## Tipos de Células

|**Célula de Código**|**Célula de Texto**|

|célula para rodas os comandos informandos na linguagem que está sendo utilizada pelo usuário.|célula feita para adicionar textos, edições em [Markdown](https://www.markdowntutorial.com), imagens, formatações em HTML5 e equações via [LaTeX](https://pt.wikipedia.org/wiki/Ajuda:Guia_de_edição/Fórmulas_TeX)|

```python
import sys
sys.path.append('assets/')
from plot_helper import *
import numpy as np
```

```python
def get_alfa(lim_inf = 0, lim_sup = 41):
    alfa = numpy.linspace(lim_inf, 2*numpy.pi, lim_sup)
    return alfa
```

```python
def coordenadas(lim_inf, lim_sup):
    coordenadas = zip(numpy.cos(get_alfa(lim_inf, lim_sup)), numpy.sin(get_alfa(lim_inf, lim_sup)))
    vetores = list(coordenadas)
    return vetores
```

```python
plot_vector(coordenadas(0, 41))
plt.title("Vetores que representam um círculo") 
```

```python
def transformacao_linear(vet):
    novosVetores = [] 
    matrizA = numpy.array([[2,1], [1,2]]) 
    for indice in range(len(vet)):
        novosVetores.append(matrizA.dot(numpy.array(vet[indice])))
    return novosVetores
```

```python
plot_vector(transformacao_linear(coordenadas(0, 41))) 
plt.title("Vetores após terem sidos transformados") 
```

```python
def giro(teta,vetores):
    matrizRotacao = numpy.array([[numpy.cos(teta), -numpy.sin(teta)],[numpy.sin(teta), numpy.cos(teta)]])
    rotacao = []  
    for indice in range(len(vetores)):
        rotacao.append(matrizRotacao.dot(numpy.array(vetores[indice]))) 
    return plot_vector(rotacao)
```

```python
teta = numpy.pi / 2 
giro(teta, transformacao_linear(coordenadas(0, 41)))
```
